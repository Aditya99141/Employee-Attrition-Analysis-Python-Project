```python
df.isna().sum()
```

This implies that there are 41, 55 & 242 values are missing from HourlyRate, MonthlyIncome & StandardHours Columns respectively.

```python
import missingno as msno
msno.bar(df)
```

Since in StandardHours column, all other values are equal to 80. We can assume that missing values would also have been 80

```python
df['StandardHours'].fillna(80,inplace=True)
```

```python
df
```

We filled missing values in StandardHours column with 80.

We can fill missing values in HourlyRate column with the average of given values.

```python
df['HourlyRate'].fillna(df['HourlyRate'].mean(),inplace=True)
```

We'll perform similar process for filling missing values in MonthlyIncome Column.

```python
df['MonthlyIncome'].fillna(df['MonthlyIncome'].mean(),inplace=True)
```

```python
df.isna().sum()
```

This implies that we have successfully filled all missing values

In the data, in SN column, nos.

```python
df.drop(['SN'], axis=1)
```

```python
df['Gender'].value_counts()
```

In the Gender column, 'M' appears 11 times, representing 'Male'. So, we will replace all instances of 'M' with 'Male'.

```python
df['Gender'].replace({'M':'Male'},inplace=True)
```

# Visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,6))
sns.countplot(x='Age', color='pink', data = df)
```

Majority of the employees are between the age of 30 and 40

```python
sns.boxplot(x='Age', y='Attrition',data=df)
```

Employees in relatively young age bracket 25-35 are more likely to leave. Hence, efforts should be made to clearly articulate the long-term vision of the company and young employees fit in that vision, as well as provide incentives in the form of clear paths to promotion for instance.


```python
sns.barplot(x="BusinessTravel", y = "Attrition", data = df)
```

The workers who travel alot are more likely to quit then other employees.

```python
plt.figure(figsize=(10,6))
sns.countplot(x='JobSatisfaction', hue="Gender", data = df)
```
